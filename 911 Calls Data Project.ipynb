## Creating new features

** In the titles column there are "Reasons/Departments" specified before the title code. These are EMS, Fire, and Traffic. Use .apply() with a custom lambda expression to create a new column called "Reason" that contains this string value.** 

**For example, if the title column value is EMS: BACK PAINS/INJURY , the Reason column value would be EMS. **

```python
#below function will create three reason category for 911 calls

df['Reason']= df['title'].apply(lambda x: x.split(':')[0])
```

```python
df['Reason'].value_counts()
#Below table shows EMS is the most common reason for the 911 calls
```

** Now let us begin to focus on time information. What is the data type of the objects in the timeStamp column? **

```python
#find datatype
#use index location to retrieve first element
type(df['timeStamp'].iloc[0])
```

** You should have seen that these timestamps are still strings. Use [pd.to_datetime](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.to_datetime.html) to convert the column from strings to DateTime objects. **

```python
#covert string to datetime using pandas
df['timeStamp'] = pd.to_datetime(df['timeStamp'])
```

Now that the timestamp column are actually DateTime objects, 
use .apply() to create 3 new columns called Hour, Month, and Day of Week.
create these columns based off of the timeStamp column

```python
df['Hour'] = df['timeStamp'].apply(lambda x:  x.hour)
df['Month'] = df['timeStamp'].apply(lambda x: x.month)
df['Day of Week'] = df['timeStamp'].apply(lambda x: x.dayofweek)
```

```python
#Day of week is stored as integer
df['Day of Week'].head(3)
```

** Notice how the Day of Week is an integer 0-6. Use the .map() with this dictionary to map the actual string names to the day of the week: **

    dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}

```python
dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
dmap
```

```python
#now use map function
df['Day of Week'] = df['Day of Week'].map(dmap)
```

** Its missing some Months, let's see if we can maybe fill in this information by plotting the information in another way, possibly a simple line plot that fills in the missing months, in order to do this, we'll need to do some work with pandas... **

** create a gropuby object called byMonth, where you group the DataFrame by the month column and use the count() method for aggregation. Use the head() method on this returned DataFrame. **

```python
#Store grouby information in new variable byMonth
#note we are only storing the count values hence count will be more or less same for each column
byMonth = df.groupby('Month').count()
byMonth.head()
```

** Now create a simple plot off of the dataframe indicating the count of calls per month. **

```python
#using pandas inbuilt visualization
#note we can use any field. I have used twp as each column in byMonth dataframe is storing count values.
#showing number of calls each month for all 12 months.
byMonth['twp'].plot()
```

** Now see if you can use seaborn's lmplot() to create a linear fit on the number of calls per month. Keep in mind you may need to reset the index to a column. **

```python
#It errors if index is not reset
#linear model plot
sns.lmplot(x='Month',y='zip',data=byMonth.reset_index())
```

**Create a new column called 'Date' that contains the date from the timeStamp column. You'll need to use apply along with the .date() method. ** 

```python
df['Date'] = df['timeStamp'].apply(lambda x: x.date())
```

** Now groupby this Date column with the count() aggregate and create a plot of counts of 911 calls.**

```python
#we could have stored it into different variable like we did earlier byMonth.
#another way of directly creating the plot
df.groupby('Date').count()['twp'].plot()
#use tight_layout function
plt.tight_layout()
```

** Now recreate this plot but create 3 separate plots with each plot representing a Reason for the 911 call**

```python
# Add filter Reason = Traffic
df[df['Reason']=='Traffic'].groupby('Date').count()['twp'].plot()
#Add title
plt.title('Traffic')
plt.tight_layout()
```

```python
#Reason = Fire

df[df['Reason']=='Fire'].groupby('Date').count()['twp'].plot()
plt.title('Fire')
plt.tight_layout()
```

```python
# Reason = EMS
df[df['Reason']=='EMS'].groupby('Date').count()['twp'].plot()
plt.title('EMS')
plt.tight_layout()
```

** Now let's move on to creating  heatmaps with seaborn and our data. We'll first need to restructure the dataframe so that the columns become the Hours and the Index becomes the Day of the Week. There are lots of ways to do this, but I would recommend trying to combine groupby with an [unstack](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.unstack.html) method. !**

```python
# Columns -->Hour
# Rows --> Day of the Week
# Analysing number of 911 calls each hour of each week day.
# Analysis will help us to identify which hour of which day are busier and vice versa

#**grouping on two columns
DayHour = df.groupby(by=['Day of Week','Hour']).count()['Reason'].unstack()
DayHour.head(5)
```

** Now create a HeatMap using this new DataFrame. **

```python
#define figuresize
plt.figure(figsize=(12,6))
sns.heatmap(DayHour,cmap='viridis')
#note more yellow means more 911 calls on that day and hour
```

** Now create a clustermap using this DataFrame. **

```python
#cluster mapping using seaborn
sns.clustermap(DayHour,cmap='viridis')
```

** Now repeat these same plots and operations, for a DataFrame that shows the Month as the column. **

```python
#cols -- Month
#Rows -- Day of Week
#Analysis of 911 calls for 12 months for each day of the week
#group by Day of week, Month
dayMonth = df.groupby(by=['Day of Week','Month']).count()['Reason'].unstack()
dayMonth.head(5)
```

```python
#Heatmap to show calls per day for each month
plt.figure(figsize=(12,6))
sns.heatmap(dayMonth,cmap='viridis')
```

```python
#cluster map
sns.clustermap(dayMonth,cmap='viridis')
```
